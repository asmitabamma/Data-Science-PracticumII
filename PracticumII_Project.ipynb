```python
# Plotting correlation matrix heatmap
df = data.select_dtypes(include=['float64', 'int64'])
correlation_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix Heatmap', fontdict={'size':15})
plt.show()
```

We only got corelation of four variables here because other fields have 'object' as data type. Upon some research, I found out that using labelEncoder would be better to convert 'object' to 'numeric' datatype. So, lets go ahead and use LabelEncoder for conversion. And then we will again generate heatmap to understand the corelation between different variables in our dataset.

```python
from sklearn.preprocessing import LabelEncoder

# Defining a threshold for the maximum number of unique values
threshold = 20 

# Identifying columns with unique values exceeding the threshold
high_cardinality_columns = [col for col in data.select_dtypes(include=['object']).columns if data[col].nunique() > threshold]

# Droping these columns from the DataFrame
df_filtered = data.drop(columns=high_cardinality_columns)

# Converting object columns to numeric using LabelEncoder in filtered dataframe
label_encoders = {}
for column in df_filtered.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    df_filtered[column] = le.fit_transform(df_filtered[column])
    label_encoders[column] = le

# Computing the correlation matrix
correlation_matrix = df_filtered.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", cbar=True)
plt.title('Heatmap of Feature Correlations')
plt.show()
```

In the code, I am using threshold because there could be many unique values for a object type column which may affect our analsis. So, limiting the values to 20 will help us identify such columns. Then, we have identified and dropped object columns with more than 20 unique values to avoid high cardinality issues. Then we are using LabelEncoder to convert the remaining categorical columns into numeric format. After preprocessing, we are using correlation matrix for the numeric columns in the filtered dataset to find the correlation between  variables.

In the heatmap, the correlation between a vast majority of the variables is shown zero which denotes a weak or no relationship between them. We can see that few cross-correlations are quite low and thus it is concluded that the attributes are independent, which means that the data could be very likely to have separate or non-intersecting properties.


# Time Series Analysis

```python
# Converting date column to datetime format
data['Timestamp'] = pd.to_datetime(data['Timestamp'])
```

```python
# Filtering out rows where 'Timestamp' is NaT (not a date)
filtered_data = data[data['Timestamp'].notna()]
filtered_data.head()
```

The time series graph shows a trend in cyberattacks from the year 2020 to 2023. During the 2020-2022 period, the number of cyberattacks remained at the same level, except for a small gain in the number of recorded cyberattacks which was the highest in 2022. This means that either the actual number of cyberattacks remained unchanged or there might have been a rise in reporting within this period of time. Nonetheless, in 2023, the number of incidents reported is very low which makes it into a remarkable change compared with previous years. The reduction may be due to increased security measures, modifications in the attack patterns, fewer reports, or other outer factors that can affect the dataset among them.

# K-Means Clustering

```python
# Importing the necessary libraries for K-means clustering analysis
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
# Converting dates to numeric

data['Timestamp'] = pd.to_datetime(data['Timestamp'])
```

```python
# Handling categorical features by encoding, Attack Type and everity Level columns are encoded into numeric values using LabelEncoder
label_encoder = LabelEncoder()
data['Attack Type'] = label_encoder.fit_transform(data['Attack Type'])
data['Severity Level'] = label_encoder.fit_transform(data['Severity Level'])
```

```python
# Selecting relevant features for clustering

features = data[[ 'Severity Level', 'Anomaly Scores', 'Packet Length']]
features.head()
```

I chose these three characteristics because they are numerically significant and capture distinct aspects of the data. In fact, severity level indicates which events have the highest priority, anomaly scores detect deviations or patterns not typically observed, and Packet Length is a size or amount of data, which can be used to classify events. These three features together can contribute to the identification of meaningful patterns and groupings in the data while ensuring the clustering focuses on relevant dimensions.

```python
print(features.isnull().sum()) # Validating no missing values are present in these selected features
```

```python
# Standardizing the features to bring them to the same scale

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
# Using Silhouette Score to find the best fit cluster number
from sklearn.metrics import silhouette_score

for n in range(2, 6):  # Test for a few cluster numbers
    kmeans = KMeans(n_clusters=n, random_state=42)
    labels = kmeans.fit_predict(scaled_features)
    score = silhouette_score(scaled_features, labels)
    print(f"Silhouette Score for n_clusters={n}: {score}")
```

It is recommended to choose the n_clusters with highest silhouette score. In our case, n_clusters=5 is comparatively higher so we will be using it for our clustering.

```python
# Performing K-Means Clustering

kmeans = KMeans(n_clusters=5, random_state=42)
data['Cluster'] = kmeans.fit_predict(scaled_features)
```

```python
# Visualizing the clusters using the first and third features

plt.figure(figsize=(10, 6))
plt.scatter(scaled_features[:, 0], scaled_features[:, 1], c=data['Cluster'], cmap='viridis')
plt.title('K-Means Clustering')
plt.xlabel('Severity Level (Scaled)')
plt.ylabel('Anomaly Scores(Scaled)')
plt.show()
```

The K-means clustering graph shows a connection between Severity Levels (x-axis) and Anomaly Scores (y-axis). Different colors indicate three clusters. The clusters are arranged vertically according to Severity Levels, demonstrating good grouping by the algorithm. Yet, within each cluster, Anomaly Scores still vary. These variations within clusters suggest that Anomaly Scores, alongside Severity Levels, reveal the extent of spread within specific severity groups. The algorithm detects some key differences among data points. Clusters based on Severity Levels emerge mainly in determining clustering structures.
This type of visualization shows how Severity Levels influence clusters and the variations in Anomaly Scores within each level. This can guide responses to anomalies by prioritizing actions based on differences in Severity Levels and Anomaly Scores.

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Apply PCA to reduce dimensions to 2 for visualization
pca = PCA(n_components=2)  # Reduce to 2 components for 2D visualization
pca_features = pca.fit_transform(scaled_features)

# Visualizing the clusters using the principal components
plt.figure(figsize=(10, 6))
plt.scatter(pca_features[:, 0], pca_features[:, 1], c=data['Cluster'], cmap='viridis', s=50)
plt.title('K-Means Clustering (PCA Visualization)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster')
plt.show()
```

The graph illustrates the final classification after implementing dimension reduction using Principal Component Analysis (PCA) on two dimensions. The four clusters (yellow, green, teal, and purple) are respectively highlighted, and their edges are very definite except in the middle, where the yellow, green, and purple clusters dramatically overlap each other. This indicates that the data points are similar to each other in some way, which in turn makes them difficult to be separated. Also, The spread along the principal components indicates the directions of maximum variance in the data, with the clusters showing moderate compactness but not perfect separation.
